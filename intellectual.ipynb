# finding highly used non article

```python
import re

pattern = re.compile("[A-Za-z]+")
answer = re.findall(pattern, book)
answer[:10]
```

```python
temp = {}
for words in answer:
    if words in temp.keys():
        temp[words] = temp[words] + 1
    else:
        temp[words] = 1
```

```python
final = [(values,keys) for (keys, values) in temp.items()]
final = sorted(final, reverse=True)
final
```
