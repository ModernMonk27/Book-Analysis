# finding a sentence in book

```python
pattern = re.compile("[A-z]{1}[^.]*thoughts[^A-Za-z]+[^.]*.")
answer = re.findall(pattern,book)
len(answer)
```

# Find most used words in the book

```python
pattern = re.compile("[A-Za-z]+")
answer = re.findall(pattern, book)
answer[:10]
```

```python
temp = {}
for words in answer:
    if words in temp.keys():
        temp[words] = temp[words] + 1
    else:
        temp[words] = 1
```

```python
temp
```

```python
final = [(values,keys) for (keys, values) in temp.items()]
sorted(final, reverse=True)
```
